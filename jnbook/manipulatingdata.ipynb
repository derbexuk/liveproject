```python
import pandas as pd
import numpy as np
```

```python
resdf = pd.read_html('https://en.wikipedia.org/wiki/NBA_All-Star_Game')[2]
```

```python
print(resdf)
```

```python
res1 = resdf.dropna(how="any")
print(res1)
```

```python
res1 = res1.drop('Game MVP', axis=1)
```

```python
res1 = res1.drop('Host arena', axis=1)
```

```python
res1.tail()
```

```python
res1['Host city'] = res1['Host city'].str.replace(",.*$", "")
```

```python
res1.head()
```

```python
res1['East'] = res1['Result'].str.extract(r"East (\d+)")
```

```python
res1['West'] = res1['Result'].str.extract(r"West (\d+)")
```

```python
res1.head()
```

```python
res1 = res1.dropna(how="any")
res1.tail()
```

```python
res1['East'] = pd.to_numeric(res1['East'])
```

```python
res1['West'] = pd.to_numeric(res1['West'])
```

```python
res1['Diff'] = abs(res1['East'] - res1['West'])
```

```python
res1.tail()
```

```python
diff_grp = res1.groupby('Diff')
```

```python
x = diff_grp.size()
```
